```python
data_location = "../Projeto-Integracao-I/Data/Real_State_Properati.csv"

df_prop = pd.read_csv(data_location, sep = '\t')
```

```python
# verificando os valores com 0 na coluna da área da propriedade "surface_total_in_m2"
mask_0 = df_prop.surface_total_in_m2 == 0
prop_null_surface = df_prop.loc[mask_0]
prop_null_surface
```

```python
# observado que as propriedades que estão valor 0 em surface_total_in_m2 são do tipo apartamento, então será verificada a media por bairro/apartamento
# mascara das propriedades do tipo apartamento
mask_apartment = df_prop['property_type'] == "apartment"
df_prop_apartment = df_prop.loc[mask_apartment]
# mascara sem as propriedades com valor 0 em surface_total_in_m2
mask_not_0 = df_prop_apartment['surface_total_in_m2'] != 0
df_prop_apartment_not0 = df_prop_apartment.loc[mask_not_0]
# verificando quanitdade de linhas
print('df_prop ', df_prop.shape, 'df_prop_apartment ', df_prop_apartment.shape,'df_prop_apartment_not0 ', df_prop_apartment_not0.shape)
# agrupamento do daframe de df_prop_apartment por bairro
df_place_mean  = df_prop_apartment_not0.groupby(['place_name'])[['surface_total_in_m2']].mean().round(2)
df_place_mean
```

```python
# mascaras por bairro que tem 0 no surface_total_in_m2 "Palermo Soho", "Recoleta", "Caballito", "Palermo Hollywood", "Barracas", "Barrio Norte"
mask_Palermo_Soho = df_place_mean.index == "Palermo Soho"
mask_Recoleta = df_place_mean.index == "Recoleta"
mask_Caballito = df_place_mean.index == "Caballito"
mask_Palermo_Hollywood = df_place_mean.index == "Palermo Hollywood"
mask_Barracas = df_place_mean.index == "Barracas"
mask_Barrio_Norte = df_place_mean.index == "Barrio Norte"
# media por bairro/apartamento
mean_Palermo_Soho = df_place_mean[mask_Palermo_Soho].surface_total_in_m2.mean()
mean_Recoleta = df_place_mean[mask_Recoleta].surface_total_in_m2.mean()
mean_Caballito = df_place_mean[mask_Caballito].surface_total_in_m2.mean()
mean_Palermo_Hollywood = df_place_mean[mask_Palermo_Hollywood].surface_total_in_m2.mean()
mean_Barracas = df_place_mean[mask_Barracas].surface_total_in_m2.mean()
mean_Barrio_Norte = df_place_mean[mask_Barrio_Norte].surface_total_in_m2.mean()
     
# comparando a media geral do surface_total_in_m2 
print('media geral surface_total_in_m2 detodos os registros:', df_prop.surface_total_in_m2.mean())
print('media geral surface_total_in_m2 somente propriedades apartamento :', df_prop_apartment.surface_total_in_m2.mean())
print('media geral surface_total_in_m2 por bairro (apartamentos) :', '\nPalermo_Soho: ', mean_Palermo_Soho,
      '\nRecoleta: ', mean_Recoleta,
      '\nCaballito: ', mean_Caballito,
      '\nPalermo_Hollywood: ', mean_Palermo_Hollywood,
      '\nPalermo_Barracas: ', mean_Barracas,
      '\nPalermo_Barrio_Norte: ', mean_Barrio_Norte
     )
```

```python
df_prop_surface_total_in_m2 = df_prop[['place_name','surface_total_in_m2']]
print(df_prop_surface_total_in_m2.shape)
print(df_prop.shape)
df_prop_surface_total_in_m2.columns
```

```python
def get_fill_value(row):    
    if row["surface_total_in_m2"] == 0:
        if row["place_name"] == "Palermo Soho":
            row["surface_total_in_m2"] = 64.12
            result = row["surface_total_in_m2"]
            
        elif row["place_name"] == "Recoleta":
            row["surface_total_in_m2"] = 127.89
            result = row["surface_total_in_m2"]
             
        elif row["place_name"] == "Caballito":
            row["surface_total_in_m2"] = 69.99
            result = row["surface_total_in_m2"]
            
        elif row["place_name"] == "Palermo Hollywood":
            row["surface_total_in_m2"] = 63.8                
            result = row["surface_total_in_m2"]
            
        elif row["place_name"] == "Barracas":
            row["surface_total_in_m2"] = 82.8   
            result = row["surface_total_in_m2"]
            
        else:
            row["surface_total_in_m2"] = 101.37 
            result = row["surface_total_in_m2"]
    else:
        result = row["surface_total_in_m2"]
    return result    

df_prop_surface_total_in_m2["surface_total_in_m2"] = df_prop_surface_total_in_m2.apply(get_fill_value, axis = 1)
```

```python
# verificando os valores com 0 na coluna da área da propriedade "surface_total_in_m2"
mask_0 = df_prop_surface_total_in_m2.surface_total_in_m2 == 0
df_prop_surface_total_in_m2_null = df_prop_surface_total_in_m2.loc[mask_0]
df_prop_surface_total_in_m2_null
```
